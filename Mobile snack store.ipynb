```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Use the "read_excel()" function from a library called "panda" that was imported as a variable called "pd" to read the excel sheet
d1_data = pd.read_excel('K-Mobile Snack Store-D1.xlsx')
d3_data = pd.read_excel('K-Mobile Snack Store-D3.xlsx')
d6_data = pd.read_excel('K-Mobile Snack Store-D6.xlsx')
d7_data = pd.read_excel('K-Mobile Snack Store-D7.xlsx')
```

```python
# Merge datasets on the 'ID' column
# This results in a single DataFrame containing all the data
merged_data = pd.merge(d1_data, d3_data, on='ID')
merged_data = pd.merge(merged_data, d6_data, on='ID')
merged_data = pd.merge(merged_data, d7_data, on='ID')
```

```python
# Display the merged dataset
print(merged_data)
```

```python
# Check for and display the number of null values in each column
print(merged_data.isnull().sum()) 
```

```python
# Drop rows with any null values
data_cleaned = merged_data.dropna()

# Now 'data_cleaned' contains the DataFrame without null values
print(data_cleaned.isnull().sum()) 
```

```python
# LIKELIHOOD OF USING
# Create a bar plot of average likelihood of using the service by gender
plt.figure(figsize=(8, 6))
data_cleaned.groupby("Q1")["Q7"].mean().plot(kind="bar", color=["#000080", "#ff99cc"])
plt.title("Average Likelihood of Using Mobile Snack Store by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Likelihood")
plt.xticks([0, 1], ["Male", "Female"])
# Show the plot
plt.show()
```

```python
# Histograms for males
plt.figure(figsize=(8, 6))
n, bins, patches = plt.hist(data_cleaned[data_cleaned["Q1"] == 1]["Q7"], bins=6, color="#000080", edgecolor="#000080")
plt.title("Likelihood of Using Mobile Snack Store (Males)")
plt.xlabel("Likelihood")
plt.ylabel("Frequency")
plt.xticks([-3, -2, -1, 0, 1, 2], ["I will not use", "Highly Unlikely", "Unlikely", "Unsure", "Likely", "Highly Likely"])

# Add data labels to the male histogram
for i, patch in enumerate(patches):
    height = patch.get_height()
    plt.text(patch.get_x() + patch.get_width() / 2, height, f"{height}", ha="center", va="bottom")
# Show the plot
plt.show()

# Histogram for females
n, bins, patches = plt.hist(data_cleaned[data_cleaned["Q1"] == 2]["Q7"], bins=6, color="#ff99cc", edgecolor="#ff99cc")  # Assign patches here
plt.title("Likelihood of Using Mobile Snack Store (Females)")
plt.xlabel("Likelihood")
plt.ylabel("Frequency")
plt.xticks([-3, -2, -1, 0, 1, 2], ["I will not use", "Highly Unlikely", "Unlikely", "Unsure", "Likely", "Highly Likely"])

# Add data labels to the female histogram
for i, patch in enumerate(patches):
    height = patch.get_height()
    plt.text(patch.get_x() + patch.get_width() / 2, height, f"{height}", ha="center", va="bottom")

plt.tight_layout()  # Adjust spacing between subplots
# Show the plot
plt.show()
```

```python
"""
Data Insights: The provided bar plot Data Visualization reveals significant gender differences in the likelihood of using the mobile snack store. Females exhibit a higher average likelihood of 3%, compared to males with an average likelihood of -1%. The data indicates that marketing should be targeted toward females, with customized products and engagement strategies such as loyalty programs and social media campaigns. 
Additionally, research into male preferences is recommended to address their lower engagement. These insights help in refining marketing and product strategies to enhance the mobile snack stores success. The data visualizations give an overview to provide a clear view of customer engagement levels by gender, aiding in strategic decision-making for marketing and product development.

"""
```

```python
#ETHNICITY 

# Define custom x-axis labels for likelihood
likelihood_labels = ["I will not use", "Highly Unlikely", "Unlikely", "Unsure", "Likely", "Highly Likely"]

# Create a dictionary to map ethnicity codes to names
ethnicity_names = {
    1: "American Indian",
    2: "African American",
    3: "Mexican American",
    4: "Other Hispanic",
    5: "Asian",
    6: "Southeast Asian",
    7: "Pacific Islander",
    8: "Filipino",
    9: "Multiple Ethnicities",
    10: "White",
    11: "Other/Not Stated",
    12: "International"
}
# 1. Bar Chart (Average)
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
merged_data.groupby("Q2")["Q7"].mean().plot(kind="bar", color="#000080")
plt.title("Average Likelihood of Using Mobile Snack Store by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Average Likelihood")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability

# Replace numerical ethnicity codes with names
plt.xticks(range(len(ethnicity_names)), [ethnicity_names[code] for code in ethnicity_names.keys()])
plt.tight_layout()  # Adjust spacing for better layout
# Show the plot
plt.show()
    
#2. pie chart
for i, ethnicity in enumerate(merged_data["Q2"].unique()):
    # Filter data for the current ethnicity
    ethnicity_data = merged_data[merged_data["Q2"] == ethnicity]
    
    # Count the number of responses for each likelihood category
    likelihood_counts = ethnicity_data["Q7"].value_counts().sort_index()
    
    # Define labels for the pie chart, ensure they match the sorted index of counts
    likelihood_labels = likelihood_counts.index
    
    # Create a pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(likelihood_counts, labels=likelihood_labels, autopct='%1.1f%%', startangle=90)
    plt.title(f"{ethnicity_names[ethnicity]} - Likelihood of Using Mobile Snack Store")
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
"""

Data Insights: The provided pie chart revelas some of the key differences in the likelihood of using mobile snack store accross different ethinic groups. For example the White had the largest proportion of respondents (29.5%) indicated a neutral response (0), while a significant portion (24.6%) showed a slight positive likelihood (1). 19.7% showed a slight negative likelihood (-1), and 11.5% each were neutral to slightly negative or slightly positive (-2 and 2) and only 3.3% showed a strong negative likelihood (-3)
The other hispanic show High Neutrality and Slightly Negative Sentimen of equal proportions of respondents (37.5%) were neutral (0) or slightly negative (-1). African American respondents were evenly distributed across all likelihood categories, with 33.3% for each neutral (0), slightly positive (1), and slightly negative (-1). The oveview of the pie chart shows These insights suggest that while White respondents are more evenly distributed with a slight inclination toward neutral and positive attitudes, Other Hispanic respondents show a tendency toward neutrality or slight negativity. In contrast, African American respondents demonstrate a balanced likelihood across all options.

"""
```

```python
# Income
# Define custom x-axis labels for likelihood
likelihood_labels = ["I will not use", "Highly Unlikely", "Unlikely", "Unsure", "Likely", "Highly Likely"]

# Create a dictionary to map income codes to names
income_names = {
    1: "$499 or less",
    2: "$500 to $999",
    3: "$1,000 to $1,499",
    4: "$1,500 to $1,999",
    5: "$2,000 or more"
}

# Plotting the box plot
plt.figure(figsize=(12, 6))
merged_data.boxplot(column="Q7", by="Q6", patch_artist=True, showmeans=True,
                    boxprops=dict(facecolor="#99ccff", color="#000080"),
                    medianprops=dict(color="#000080"),
                    capprops=dict(color="#000080"),
                    whiskerprops=dict(color="#000080"),
                    meanprops=dict(markerfacecolor="#ff5733", marker='D', markeredgecolor='#000080'))

# Replace numerical income codes with names
plt.xticks(range(1, len(income_names) + 1), [income_names[code] for code in sorted(income_names.keys())], rotation=45, ha="right")

# Adding title and labels
plt.title("Average Likelihood of Using Mobile Snack Store by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Likelihood")
plt.suptitle('')  # Remove the automatic 'Q7 by Q6' title

plt.tight_layout()
plt.show()

# 2. Box Plots (Combined)
plt.figure(figsize=(12, 6))
merged_data.boxplot(column="Q7", by="Q6", patch_artist=True,
            boxprops=dict(facecolor="#99ccff", color="#000080"),
            medianprops=dict(color="#000080"),
            capprops=dict(color="#000080"),
            whiskerprops=dict(color="#000080"))
plt.title("Likelihood of Using Mobile Snack Store by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Likelihood")
plt.xticks(range(1, len(income_names) + 1), [income_names[code] for code in sorted(income_names.keys())], rotation=45, ha="right")
# Show the plot
plt.show()

#3. Iterate over unique income levels and plot box plots for each
for i, income in enumerate(merged_data["Q6"].unique()):
    # Filter data for current income level
    income_data = merged_data[merged_data["Q6"] == income]
    
    plt.figure(figsize=(8, 6))
    plt.boxplot(income_data["Q7"], patch_artist=True, showmeans=True,
                boxprops=dict(facecolor="#99ccff", color="#000080"),
                medianprops=dict(color="#000080"),
                capprops=dict(color="#000080"),
                whiskerprops=dict(color="#000080"),
                meanprops=dict(markerfacecolor="#ff5733", marker='D', markeredgecolor='#000080'))

    # Set custom x-axis labels for likelihood
    plt.title(f"{income_names[income]} - Box Plot")
    plt.xlabel("Likelihood")
    plt.ylabel("Frequency")
    plt.xticks([1], [income_names[income]])
    plt.xticks(range(1, len(likelihood_labels) + 1), likelihood_labels)

    # Show the plot
    plt.show()
    
```

```python
"""
Data Insights: The above box plot reveals the likelihood of using a mobile snack store across different income levels. Low-Income Levels show high variability and generally lower median likelihoods, indicating mixed opinions and a less favorable view towards using the mobile snack store.
where as the Middle-Income Levels respondents exhibit moderate variability with medians closer to neutral, suggesting a balanced or slightly positive stance and the High-Income Levels groups have lower variability and higher median likelihoods, indicating a more uniform and favorable attitude towards using the service.
The overall trend is Higher income levels generally show a stronger inclination to use the mobile snack store, with consistent mean and median alignment.

"""
```
